```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score


file = np.loadtxt("eucalyptus.txt")
h, c = np.loadtxt("eucalyptus.txt", unpack=True)
c = np.reshape(c, (-1, 1))

print("h : ", h)
print("c : ", c)
plt.scatter(c, h)

reg_eucalyptus = LinearRegression().fit(c, h)
y_pred = reg_eucalyptus.predict(c)
scores = cross_val_score(reg_eucalyptus, c, h, cv=5)
print("scores par validation croisée : ",scores)
print("la moyenne du tableau scores : ",scores.mean())
print("la valeur de hauteur predite pour l’eucalyptus de circonference 22.8 : ", reg_eucalyptus.predict([[22.8]]))

plt.plot(c, y_pred, "r")
plt.show()
```

```python
sqrt_c = np.sqrt(c)
concat_c = np.concatenate((c, sqrt_c), axis=1)
reg_eucalyptus3D = LinearRegression().fit(concat_c, h)
y_pred3D = reg_eucalyptus3D.predict(concat_c)
scores3D = cross_val_score(reg_eucalyptus3D, concat_c, h, cv=5)
print("scores par validation croisée : ",scores3D)
print("la moyenne du tableau scores : ",scores3D.mean())
```

```python
from mpl_toolkits.mplot3d import Axes3D
result = reg_eucalyptus3D.predict(concat_c)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(c, h,result)
ax.plot(c, h, result,'r')
```
