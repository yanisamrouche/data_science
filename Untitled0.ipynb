```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

X = np.array([[5.5],[6.0],[6.5],[6.0],[5.0],[6.5],[4.5],[5]])
y = [420,380,350,400,440,380,450,420]

reg = LinearRegression().fit(X, y)
reg_interfalse = LinearRegression(fit_intercept=False).fit(X, y)
 
y_pred = reg.predict(X)
y_pred_interfalse = reg_interfalse.predict(X)

print("l’erreur quadratique moyenne ",mean_squared_error(y, y_pred))
print("coef : ",reg.coef_)
print("intercept : ",reg.intercept_)
print("le score : ", reg.score(X, y)) 
plt.plot(X, y_pred, "r")
plt_y = []
for i in X:
    y_ = reg.coef_*i + reg.intercept_
    plt_y.append(y_)
    
plt.plot(X, plt_y, "g")
plt.scatter(X,y)
plt.show()
### intercept = false

print("l’erreur quadratique moyenne (interfalse) ",mean_squared_error(y, y_pred_interfalse))
print("coef (interfalse) :  ",reg_interfalse.coef_)
print("intercept (interfalse) :  ",reg_interfalse.intercept_)
print("le score (interfalse) : ", reg_interfalse.score(X, y)) 
plt.plot(X, y_pred_interfalse, "r")
plt_y_interfalse = []
for i in X:
    y_ = reg_interfalse.coef_*i + reg_interfalse.intercept_
    plt_y_interfalse.append(y_)
    
plt.plot(X, plt_y_interfalse, "g")
plt.scatter(X,y)
plt.show()

```
